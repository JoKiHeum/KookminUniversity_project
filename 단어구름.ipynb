# 단어구름

```python
from wordcloud import WordCloud
import pandas as pd
```

```python
word_count_kiwi = pd.read_excel('word_count_kiwi.xlsx',index_col=0)
```

```python
count_dic_kiwi = dict(zip(word_count_kiwi['단어'], word_count_kiwi['빈도']))
```

```python
count_dic_kiwi = word_count_kiwi.set_index('단어')['빈도'].to_dict()
```

```python
wc = WordCloud(font_path='C:/Windows/Fonts/NanumGothic.ttf',
               background_color='white',
               max_words=50,
               width=400, height=300)
```

```python
cloud_kiwi = wc.generate_from_frequencies(count_dic_kiwi)
```

```python
cloud_kiwi.to_image()
```
